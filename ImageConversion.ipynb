```python
import os
from PIL import Image, ImageDraw , ImageFont
import io
```

# From Pillow to bytes

```python
def pillow_to_bytes(image:Image):
    imgByteArr = io.BytesIO()
    image.save(imgByteArr, format=image.format)
    imgByteArr = imgByteArr.getvalue()
    return imgByteArr
```

```python
pil_img = Image.open("dataset/povtorka/real_00001.jpg")
type(pil_img)
```

```python
byte_img = pillow_to_bytes(pil_img)
type(byte_img)
```

# From Bytes to Pillow

```python
def bytes_to_pillow(byte: bytes):
    pillow = Image.open(io.BytesIO(byte))
    return pillow
```

```python
type(byte_img)
```

```python
pil = bytes_to_pillow(byte_img)
type(pil)
```
