```python
from PIL import Image, ImageDraw , ImageFont
```

```python
board = Image.open('img/b.jpg')
survived = Image.open('img/survived.jpeg')
```

```python
img1 = Image.open('dataset/povtorka/real_00001.jpg')
img2 = Image.open('dataset/povtorka/real_00002.jpg')
img3 = Image.open('dataset/povtorka/real_00003.jpg')
img4 = Image.open('dataset/povtorka/real_00004.jpg')
img5 = Image.open('dataset/povtorka/real_00005.jpg')
type(img1)
```

```python
def verdict(img:Image, povtorka:bool):
    if povtorka:
        width, height = img.size 
        img = img.resize((int(width/2), int(height/2)))
        board.paste(img,(100,150))
        font = ImageFont.truetype("Arial.ttf", size = 50)
        draw = ImageDraw.Draw(board)
        draw.text((65,500), "Повторний Курс", fill = 'black',font = font)
        return board
    else:
        width, height = img.size 
        img = img.resize((int(width/1.5), int(height/1.5)))
        survived.paste(img,(200,130))
        font = ImageFont.truetype("Arial.ttf", size = 75)
        draw = ImageDraw.Draw(survived)
        draw.text((250,550), "Survived", fill = 'green',font = font)
        return survived
verdict(img5,2)
```

```python
def frame_only(img:object, povtorka:bool):
    if povtorka:
        width, height = img.size 
        img = img.resize((int(width/2), int(height/2)))
        board.paste(img,(100,150))
        return board
    else:
        width, height = img.size 
        img = img.resize((int(width/1.5), int(height/1.5)))
        survived.paste(img,(200,130))
        return survived
frame_only(img2, 1)
```

```python
def text_only(img:object, povtorka:bool)-> object:
    if povtorka:
        font = ImageFont.truetype("Arial.ttf", size = 50)
        draw = ImageDraw.Draw(img)
        draw.text((100,150), "Повторний Курс", fill = 'red',font = font)
        return img
    else:
        font = ImageFont.truetype("Arial.ttf", size = 75)
        draw = ImageDraw.Draw(img)
        draw.text((170,140), "Survived", fill = 'green',font = font)
        return img
text_only(img3,1)
```
